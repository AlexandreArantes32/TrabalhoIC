## pegar os frames do vídeo

```python
cap = cv2.VideoCapture('/content/drive/MyDrive/TrabalhoIC/vídeos/document_5089348937941254835.mp4')
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    frame_filename = f"frame_{frame_count:04d}.jpg"
    cv2.imwrite(os.path.join(dataset_path , frame_filename), frame)

    frame_count += 1

cap.release()
cv2.destroyAllWindows()
```
